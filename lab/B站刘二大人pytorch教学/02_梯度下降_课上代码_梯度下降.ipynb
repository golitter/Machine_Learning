```python
import numpy as np 
import matplotlib.pyplot as plt
```

梯度下降是机器学习中的常用算法，通过不断迭代计算函数的梯度，判断该点的某一方向和目标之间的距离，最终求得最小的损失函数和相关参数，为建立线性模型提供支持。<br>
https://blog.csdn.net/qq_45066628/article/details/123761421

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0
cost_list = []
def forward(x): # x * k^
    return x * w

def cost(xs, ys):
    '''
    cost函数为  ( l / len(xs)) * 累计 (y_pred - y)**2
                                    ( (w * x) - y ) ** 2
    '''
    cost = 0
    for x,y in zip(xs, ys):
        y_pred = forward(x)
        cost += (y_pred - y) ** 2
    return cost / len(xs)

def gradient(xs, ys):
    '''
    z = b**2 ; b = (a - y); a = w * x
    z = 2 * b * w == 2 * w * (w * x - y)
    '''
    grad = 0
    for x,y in zip(xs, ys):
        grad += 2 * x * ( x * w - y)
    return grad / len(xs)

print("Predict (before training)", 4, forward(4))
for epoch in range(100): # 迭代100次，epoch 表示当前的迭代次数
    cost_val = cost(x_data, y_data)
    grad_val = gradient(x_data, y_data)
    w -= 0.01 * grad_val
    print("Epoch: ",epoch, "w = ", w, "loss = ", cost_val)
    cost_list.append(cost_val)
print("Predict (after training)", 4, forward(4))
```

```python
plt.plot(range(1,101), cost_list)
plt.ylabel("cost")
plt.xlabel("epoch")
plt.show()
```
