```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 4.0]
y_data = [7.0, 10.0, 16.0]

w_list = []
b_list = []
mse_list = []

def forward(x):
    return w*x + b

def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

sum_min = 1e9
w_opt = 0
b_opt = 0

for w in np.arange(0.0, 4.0, 0.1):
    for b in np.arange(0.0, 4.0, 0.1):
        sum = 0
        for x_val, y_val in zip(x_data, y_data):
            y_pred_val = forward(x_val)
            loss_val = loss(x_val, y_val)
            sum += loss_val
            print('\t', x_val, y_val, y_pred_val, loss_val)
        print("MSE = ", sum/3)
        w_list.append(w)
        b_list.append(b)
        mse_list.append(sum/3)

        if(sum_min > sum):
            w_opt = w
            b_opt = b

```

```python

print("y = {:.2f}*x + {:.2f}".format(w_opt, b_opt))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# [x, y] = np.meshgrid(w_list, b_list)
x = w_list
y = b_list
z = mse_list
ax.plot(x,y,z)
plt.show() # 三维图效果不好
```
