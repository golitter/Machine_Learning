### 尝试不同的激活函数 activate function
只需要将模型里的激活函数改一下即可。
`self.sigmoid = torch.nn.Sigmoid()` ==> `self.activate = torch.nn.ReLU()`即可。<br>
同时，注意不同的激活函数的值域范围，可能在最后一层的激活函数需要用连续的激活函数等。<br>

```python
# 使用 ReLU() 激活函数

# design model using class
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1 = torch.nn.Linear(8, 6)
        self.linear2 = torch.nn.Linear(6, 4)
        self.linear3 = torch.nn.Linear(4, 1)
        self.activate = torch.nn.ReLU() # 改了激活函数

    def forward(self, x):
        x = self.activate(self.linear1(x))
        x = self.activate(self.linear2(x))
        x = self.activate(self.linear3(x)) # 可能最后一个激活函数换成其他的激活函数
        ''' 
            ReLU激活函数在x = 0时没有定义，
            导致y出来可能有不同，最后一个激活函数可以用其他激活函数进行替代
        '''
        return x
```
